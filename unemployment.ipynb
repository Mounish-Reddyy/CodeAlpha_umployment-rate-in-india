```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
print(df.head())
```

```python
print(df.info())
```

```python
df = df.dropna()
```

```python
df.columns = df.columns.str.strip()
```

```python
print(df.columns)
```

```python
df.groupby('Region')['Estimated Unemployment Rate (%)'].agg(['mean', 'max', 'min', 'std'])
```

```python
sns.barplot(
    x='Region',
    y='Estimated Unemployment Rate (%)',
    data=df
)

plt.xticks(rotation=90)
plt.title("State Wise Unemployment Rate")
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.lineplot(
    x='Date',
    y='Estimated Unemployment Rate (%)',
    data=df
)

plt.title("Unemployment Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.heatmap(df.corr(numeric_only=True), annot=True)

plt.title("Correlation Heatmap")
plt.show()
```

```python
_period = df[df['Date'] >= '2020-03-01']

print("\nAverage Unemployment During Covid:")
print(df['Estimated Unemployment Rate (%)'].mean())
```
